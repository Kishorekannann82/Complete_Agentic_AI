```python
import os
from dotenv import load_dotenv
load_dotenv()
GROQ_API_KEY = os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq
llmModel = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)
```

```python
response = llmModel.invoke("What is LangGraph?")

print(response.content)
```

```python
## Other models like open ai api keyy
from langchain_openai import OpenAI
llmModel = OpenAI()
response  = llmModel.invoke("Tell me one fun fact about the Kennedy family.")
print(response.content)
```

```python
for chunk in llmModel.stream(
    "Tell me one fun fact about the Kennedy family.i need detail response"
):
    print(chunk, end="", flush= True)
```
